```python
topic modeling, order not matter, what topis is talked about in the document
 input: doccument-term matrix, start with what the bags of words are 
 gensim: toolkit used for topic modeling , used Latent Dirichlet Allocation (LDA)
 we use nltk to do some part-of-speech tagging 
 output: goal is to find out themes of each topic / document
        
every document consists of a mix of topics 
every topic consists of a mix of words 

latent dirichlet allocation 

input: document-term matrix, numbere of topics, number of iterations 
gensim: go through the process of finding words / topic distributions
output: the top words in each topic , human job to interpret the output
this is a probabilistic approach to topic modeling 

    
```

```python
# topic modeling - attemp #1 (all text)

# read in our document-term matrix 
import pandas as pd 
import pickle 

data = pd.read_pickle('dtm_stop.pkl')
data
```

```python
# import the necessary modules for LDA with gensim 
# terminal / anaconda navigator: conda install -c conda-forge gensim 
from gensim import matutils, models
import scipy.sparse 

import logging 
logging.basicConfig(format='%(asctime)s : %(levelname)s :%(message)s', level=logging.INFO)
```

```python
# one of the required inputs is a term-document matrix 
tdm = data.transpose()
tdm.head()
```

```python
# we are going to put the term-document matrix into a new gensim form format 
# from df --> sparse matrix 
sparse_counts = scipy.sparse.csr_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
```

```python
# Gensim also requires dictionary of the all terms and their respective location 
# in the term-document matrix 
cv = pickle.load(open("cv_stop.pkl", "rb"))
id2word = dict((v,k) for k, v in cv.vocabulary_.items())
```

we have the corpus (term-document matrix) and id2word (dictionary of location: term) and id2word (dictionay of location: term), we need to 
specify two other parameters - the number of topics and the number of passes, let's start with topics at 2, see if results make sense, then increase number from there

LDA requires a term document matrix, convert it into a sparrse matrix first, then convert it into a specific gensim corpus, which gensim can read in 

gensim also requires a dictionary of each term and their location in that document term matrix, which you can get it from count vectorizer 

```python
# now that we have the corpus (term-document matrix) and id2word (dictionary of location: term)
# we need to specify two other parameters as well - the number of topics 
# and the number of  of passes 
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=2, passes=10)
lda.print_topics()
```

```python
topic modeling - attemp #2 (nouns only)
one popular trick is to look only at terms that aree from one part of speech 
only nouns, only adjectives, etc..., check the Upenn tag set : 
https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html
```

```python
# let's create a function to pull out nouns from a string of text 
from nltk import word_tokenize, pos_tag 

def nouns(text):
    '''given a string of text, tokenize the next and pull out only the bonus'''
    is_noun = lambda pos: pos[:2] == 'NN'
    tokenized = word_tokenize(text)
    all_nouns = [word for (word, pos) in pos_tag(tokenized) if is_noun(pos)]
    return ' '.join(all_nouns)
```

```python
# Read in the cleanedd data, before the CountVectorizer step 
data_clean = pd.read_pickle('data_clean.pkl')
data_clean 
```

```python
# run below to fix this error - Resource punkt not found.
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# apply the nouns function to the transcript to filter only on nouns 
data_nouns = pd.DataFrame(data_clean.transcript.apply(nouns))
data_nouns
```

```python
# create a new document-term matrix using only nouns 
from sklearn.feature_extraction import text 
from sklearn.feature_extraction.text import CountVectorizer  

# Re-add the additional stop words since we are recreating the document-term document 
add_stop_words = ['like', 'im', 'know', 'just', 'dont', 'thats', 'right', 'people',
                 'youre', 'got', 'gonna', 'time', 'think', 'yeah', 'said']
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)
# recreate a document-term matrix with only nouns 
cvn = CountVectorizer(stop_words=stop_words)
data_cvn = cvn.fit_transform(data_nouns.transcript)
data_dtmn = pd.DataFrame(data_cvn.toarray(), columns=cvn.get_feature_names_out())
data_dtmn.index = data_nouns.index 
data_dtmn 
```

```python
# create the gensim corpus 
corpusn = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmn.transpose()))

# create the vocabulary dictionary 
id2wordn = dict((v, k) for k, v in cvn.vocabulary_.items())
```

```python
# let's start with 2 topics 
ldan = models.LdaModel(corpus=corpusn, num_topics=2, id2word=id2wordn, passes=10)
ldan.print_topics()
```

```python
# let's try topcis = 3 
ldan = models.LdaModel(corpus=corpusn, num_topics=3, id2word=id2wordn, passes=10)
ldan.print_topics()
```

```python
# let's try topcis = 3 
ldan = models.LdaModel(corpus=corpusn, num_topics=4, id2word=id2wordn, passes=10)
ldan.print_topics()
```

topic modeling - attemp #3 (nouns and adjectives)


```python
# let's create a function to pull out nouns and adjectives from a string of text 
def nouns_adj(text):
    '''given a string of text, tokenize the text and pull out only the nouns and adjectives'''
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)]
    return ' '.join(nouns_adj) 
    
```

```python
# apply the nouns functionn to the transcript to filter only on nouns 
data_nouns_adj = pd.DataFrame(data_clean.transcript.apply(nouns_adj))
data_nouns_adj 
```

```python
# apply the nouns function to the transcripts to filter only on nouns 
data_nouns_adj = pd.DataFrame(data_clean.transcript.apply(nouns_adj))
data_nouns_adj 
```

```python
# create a new document-term matrix using only nouns and adjectives, also removing common words 
cvna = CountVectorizer(stop_words=stop_words, max_df=.8)
data_cvna = cvna.fit_transform(data_nouns_adj.transcript)
data_dtmna = pd.DataFrame(data_cvna.toarray(), columns=cvna.get_feature_names_out())
data_dtmna.index = data_nouns_adj.index 
data_dtmna
```

```python
# create the gensim corpus 
corpusna = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmna.transpose()))

# create the vocabulary dictionary 
id2wordna = dict((v,k) for k, v in cvna.vocabulary_.items())
```

```python
# let's startt with 2 topics 
ldana = models.LdaModel(corpus=corpusna, num_topics=2, id2word=id2wordna, passes=10)
ldana.print_topics()
```

```python
# let's startt with 3 topics 
ldana = models.LdaModel(corpus=corpusna, num_topics=3, id2word=id2wordna, passes=10)
ldana.print_topics()
```

```python
# let's startt with 4 topics 
ldana = models.LdaModel(corpus=corpusna, num_topics=4, id2word=id2wordna, passes=10)
ldana.print_topics()
```
