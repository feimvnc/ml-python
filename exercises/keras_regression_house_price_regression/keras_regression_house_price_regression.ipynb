```python
# create model and deal with overfitting
X = df.drop('diagnosis',axis=1).values
y = df['diagnosis'].values
```

```python
# X
# y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.25, random_state=42)
```

```python
# normalize data 
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# create model 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
# create a model
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
#binary classification
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
# train the model
model.fit(x=X_train,y=y_train,epochs=300,validation_data=(X_test,y_test))
```

```python
#plot loss
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()  
# validation worse and worse, overfitting
# we can try early stopping, refine model
```

```python
 # create a model
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
#binary classification
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# help(EarlyStopping) 
# Quantity to be monitored 
# patience: no improvement after which training will be stopped
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
# tracking val_loss
# min: when quantity monitored has stopped decreasing
# verbose: more output
# wait 25 epochs even after we detected a stopping point
```

```python
# epochs, can choose a large number, callbacks early_stop can help stop the training
model.fit(x=X_train,y=y_train,epochs=300,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
# Epoch 00122: early stopping
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()  # this is the graph we want to see
```

```python
# another option we can use to prevent overfitting 
# is to use drop out layers
from tensorflow.keras.layers import Dropout
```

```python
 # create a model
model = Sequential()
model.add(Dense(30,activation='relu'))
# 0 = no neuron dropout
# 1 = all neuron dropout
# 0.5 = half neurons in each batch are turned off
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))   # randomly selected
#binary classification
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
# use previous training along with early_stop
model.fit(x=X_train,y=y_train,epochs=300,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()   # both training/validation loss are going down
```

```python
model.predict_classes(X_test)
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
